````python
def fib():
  n = int(input("Enter the number of terms: "))
  a = 0
  b = 1
  print(a)
  print(b)
  for i in range(n - 2):
    c = a + b
    a = b
    b = c
    print(c)
fib()
````

````python
def myPow(x, n):
  result = 1
  i = 0
  while i < n:
    result *= x
    i += 1
  return result
print(myPow(4, 4))
````
